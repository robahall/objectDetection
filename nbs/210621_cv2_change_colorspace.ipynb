```python
from pathlib import Path
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ROOT_DIR = Path('/home/rob/PycharmProjects/objectDetection')
DATA_DIR = ROOT_DIR /'data'
```

```python
img1 = cv2.imread(str(DATA_DIR / 'images' / 'trafficheader.jpg'))[:,:,::-1]
#img1 = cv2.resize(img1, (1200,800))
img2 = cv2.imread(str(DATA_DIR / 'images' / 'lamborghini.jpg'))[:,:,::-1]
#img2 = cv2.resize(img2, (1200,800))    
```

```python
## Define color space
# Create black image 10x1
im = np.zeros([1,10,3], dtype=np.uint8)

# Fill with colours of rainbow and greys
im[0,0,:]=[255,0,0]       # red
im[0,1,:]=[255,165,0]     # orange
im[0,2,:]=[255,255,0]     # yellow
im[0,3,:]=[0,255,0]       # green
im[0,4,:]=[0,0,255]       # blue
im[0,5,:]=[75,0,130]      # indigo
im[0,6,:]=[238,130,238]   # violet
im[0,7,:]=[0,0,0]         # black
im[0,8,:]=[127,127,127]   # grey
im[0,9,:]=[255,255,255]   # white

hsv=cv2.cvtColor(im,cv2.COLOR_RGB2HSV)
print(hsv)
```

```python
orange = np.uint8([[[255,165,0]]])
hsv_orange = cv2.cvtColor(orange, cv2.COLOR_BGR2HSV)
print(hsv_orange)
```

## Change Color-space

```python
# Convert BGR to HSV
hsv = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)

# define range of orange color in HSV
lower_orange = np.array([101-10,175,50])
upper_orange = np.array([101+10,255,255])

# threshold the hsv image to get only orange colors
mask = cv2.inRange(hsv, lower_orange, upper_orange)

# threshold the hsv image to get only orange colors
res = cv2.bitwise_and(img1, img1, mask=mask)
```
